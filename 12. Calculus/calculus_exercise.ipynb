1. Plot the absolute value of x and its derivative. Choose an interval to have a nice looking shape.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import derivative
```

```python
x = np.linspace(1, 100, 10)

x = -(-x)
y = x  

plt.plot(x, y)     
```

2. Plot the function $x\log(x)$ and its derivative.

```python
def log(x):
    return x*np.log(x)
x = np.linspace(1, 100, 10)

x_log_derivative = derivative(log, x, dx=1e-6)
plt.plot(x, log(x))
```

```python
plt.plot(x, x_log_derivative)
```

3. Plot the function $e^{x}\log(x)+5$ and its derivative.

```python
def log(x):
    return ((np.exp(x)*np.log(x)) + 5)
x = np.linspace(1, 100, 10)
plt.plot(x, log(x))
```

```python
e_x_log_derivative = derivative(log, x, dx=1e-6)
plt.plot(x, e_x_log_derivative)
```
