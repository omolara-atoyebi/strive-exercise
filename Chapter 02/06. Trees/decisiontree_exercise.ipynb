```python
import pandas as pd 
import numpy as np
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import plot_confusion_matrix
import matplotlib.pyplot as plt
```

```python
iris_data = load_iris()
iris_data.keys()
```

```python
iris_data.target_names
```

```python
df = pd.DataFrame(np.c_[iris_data['data'], iris_data['target']], columns = iris_data['feature_names']+['target_names'])
df
```

```python
df.dtypes
```

```python
X,y = df.iloc[:,:-1], df.iloc[:,-1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 45 , test_size = 0.3)
```

```python
y_train
```

```python
dt_clf = DecisionTreeClassifier(random_state = 45)
dt_clf = dt_clf.fit(X_train, y_train)
```

```python
plt.figure(figsize = (18,16))
plot_tree(dt_clf, filled = True , rounded = True,
          class_names = ['setosa', 'versicolor', 'virginica'], feature_names = X.columns);
```

```python
plot_confusion_matrix(dt_clf, X_test,y_test, display_labels= ['setosa', 'versicolor', 'virginica'])
```
